#### You can see all the columns in all the datasets is the same so we can concatenate all the dataset into one by creating separate columns for weekdays/weekend and cities.

```python
 #creating a function to create two new columns for weekdays/weekend and cities and merging them. 
def Combine_same_city(file1,WD,file2,WE,city):
    file1['week_type'] = WD
    file2['week_type'] = WE
    file1.drop(columns = ['Unnamed: 0'],inplace=True) #droping the unnecessary column in the dataset
    file2.drop(columns = ['Unnamed: 0'],inplace=True) ##droping the unnecessary column in the dataset
    combined = pd.concat([file1,file2])
    combined['city'] = city
    return combined
```

```python
#combining dataset as per cities
amsterdam = Combine_same_city(amsterdam_WD,'weekday',amsterdam_WE,'weekend',"Amsterdam")
athens = Combine_same_city(athens_WD,'weekday',athens_WE,'weekend',"Athens")
barcelona = Combine_same_city(barcelona_WD,'weekday',barcelona_WE,'weekend',"Barcelona")
berlin = Combine_same_city(berlin_WD,'weekday',berlin_WE,'weekend',"Berlin")
budapest = Combine_same_city(budapest_WD,'weekday',budapest_WE,'weekend',"Budapest")
lisbon = Combine_same_city(lisbon_WD,'weekday',lisbon_WE,'weekend',"Lisbon")
london = Combine_same_city(london_WD,'weekday',london_WE,'weekend',"London")
paris = Combine_same_city(paris_WD,'weekday',paris_WE,'weekend',"Paris")
rome = Combine_same_city(rome_WD,'weekday',rome_WE,'weekend',"Rome")
vienna = Combine_same_city(vienna_WD,'weekday',vienna_WE,'weekend',"Vienna")
```

```python
#taking all the city wise data into a list for concatination
city_data = [amsterdam,athens,barcelona,berlin,budapest,lisbon,london,paris,rome,vienna]
```

```python
#Concatinating all the city wise data into a single file (appending all the rows )
europe_data = pd.concat(city_data, ignore_index = True)
```
